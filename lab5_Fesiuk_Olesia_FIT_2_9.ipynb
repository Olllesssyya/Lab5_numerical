```python
import numpy as np
from scipy import optimize
import math

x0 = -0.07
y0 = 0.5
delta = 0.1

def f1(y):
    return math.sin(y+1) - 1 #наша функція
def f2 (x):
    return 1 - 1/2*math.cos(x) #наша функція
def f3(x): #Задаємо функцію для перевірки
    return -x[0] +  math.sin(x[1]+1)  - 1, 2*x[1] + math.cos(x[0]) - 2
def iter (x,y,e):
    xn = x
    yn = y
    xn1 = f2(x)
    yn1 = f1(y)
    n = 1
    while ((abs(xn1-xn)>=e) & (abs(yn1-yn) >=e)):
        xn = xn1
        yn = yn1
        xn1 = f2(yn)
        yn1 = f1(xn)
        n += 1
    print ('Проста ітерація: \nx=', xn, '\ny=',yn,'\nКількість ітерацій = ',n)

iter(x0,y0,0.0001)
s = optimize.root(f3, [0.,0.], method = 'hybr') #Перевірка розв*язку Відповіді мають співпасти
print ('Виконуємо перевірку',s.x)
```
